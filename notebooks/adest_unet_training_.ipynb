```python
import os
import random
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import importlib.util
import cv2
import gc
from tensorflow.keras.callbacks import EarlyStopping

# === DICOMDataGen ===
class DICOMDataGen(tf.keras.utils.Sequence):
    def __init__(self, image_ids, img_path, mask_path, batch_size=4, image_size=512, shuffle=True):
        self.ids = image_ids
        self.img_path = img_path
        self.mask_path = mask_path
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.on_epoch_end()

    def __load__(self, item):
        image = cv2.imread(os.path.join(self.img_path, item), cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
        mask = cv2.imread(os.path.join(self.mask_path, item), cv2.IMREAD_GRAYSCALE).astype(np.float32)
        mask = (mask >= 255).astype(np.float32)

        image = cv2.resize(image, (self.image_size, self.image_size), interpolation=cv2.INTER_LINEAR)
        mask = cv2.resize(mask, (self.image_size, self.image_size), interpolation=cv2.INTER_NEAREST)

        return np.expand_dims(image, axis=-1), np.expand_dims(mask, axis=-1)

    def __getitem__(self, index):
        batch_ids = self.ids[index * self.batch_size : (index + 1) * self.batch_size]
        images, masks = zip(*(self.__load__(i) for i in batch_ids))
        return np.array(images), np.array(masks)

    def on_epoch_end(self):
        if self.shuffle:
            random.shuffle(self.ids)
        gc.collect()

    def __len__(self):
        return int(np.ceil(len(self.ids) / self.batch_size))


# === Metrics and Loss Functions ===
def dice_coef(y_true, y_pred):
    smooth = 1e-6
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1. - dice_coef(y_true, y_pred)

def iou_score(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    union = tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)


# === Compute Shape Prior ===
def compute_shape_prior_from_generator(generator):
    all_masks = []
    for i in range(len(generator)):
        _, masks = generator[i]
        all_masks.append(masks)
    all_masks = np.concatenate(all_masks, axis=0)
    prior = np.mean(all_masks, axis=0)  # shape: (H, W, 1)
    return prior


# === Custom Total Loss ===
def total_loss_fn(prior):
    prior_tensor = tf.convert_to_tensor(prior, dtype=tf.float32)
    def loss(y_true, y_pred):
        d_loss = dice_loss(y_true, y_pred)
        s_loss = tf.reduce_mean(tf.square(y_pred - prior_tensor))
        return d_loss + 0.1 * s_loss
    return loss
```

```python
import os
import shutil

def clean_folder(folder):
    deleted = 0
    for root, dirs, files in os.walk(folder):
        # Delete .ipynb_checkpoints folders
        if ".ipynb_checkpoints" in dirs:
            checkpoint_path = os.path.join(root, ".ipynb_checkpoints")
            shutil.rmtree(checkpoint_path)
            print(f"Deleted: {checkpoint_path}")
            deleted += 1
        # Delete hidden files like .DS_Store or others
        for file in files:
            if file.startswith('.') or file.lower() in ['thumbs.db']:
                filepath = os.path.join(root, file)
                os.remove(filepath)
                print(f"Deleted: {filepath}")
                deleted += 1
    return deleted

# Define your paths
paths = [
    "/workspace/final_split_balanced/final_split_balanced/augmentedtrain/images",
    "/workspace/final_split_balanced/final_split_balanced/augmentedtrain/masks",
    "/workspace/final_split_balanced/final_split_balanced/valid/images",
    "/workspace/final_split_balanced/final_split_balanced/valid/masks"
]

# Clean all
total_deleted = 0
for path in paths:
    print(f"\nCleaning: {path}")
    total_deleted += clean_folder(path)

print(f"\n✅ Cleanup complete. Total items deleted: {total_deleted}")
```

```python
# ===========================
# Hyperparameters
# ===========================
img_size = 512
batch_size = 4
epochs = 50
lr = 1e-4
filters = [32, 64, 128, 256, 512]

import os
import numpy as np
import tensorflow as tf
import importlib.util
import matplotlib.pyplot as plt
import cv2
import random
import gc
from tensorflow.keras.callbacks import EarlyStopping

# ===========================
# Load DualEncoderUNet dynamically
# ===========================
dual_encoder_path = "/workspace/DualEncoderUNet.py"
spec = importlib.util.spec_from_file_location("DualEncoderUNet", dual_encoder_path)
dual_module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(dual_module)
build_dual_encoder_unet = dual_module.build_dual_encoder_unet

# ===========================
# Data Generator
# ===========================
class DICOMDataGen(tf.keras.utils.Sequence):
    def __init__(self, image_ids, img_path, mask_path, batch_size=batch_size, image_size=img_size, shuffle=True):
        self.ids = image_ids
        self.img_path = img_path
        self.mask_path = mask_path
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.on_epoch_end()

    def __load__(self, item):
        image = cv2.imread(os.path.join(self.img_path, item), cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
        mask = cv2.imread(os.path.join(self.mask_path, item), cv2.IMREAD_GRAYSCALE).astype(np.float32)
        mask = (mask >= 255).astype(np.float32)
        image = cv2.resize(image, (self.image_size, self.image_size), interpolation=cv2.INTER_LINEAR)
        mask = cv2.resize(mask, (self.image_size, self.image_size), interpolation=cv2.INTER_NEAREST)
        return np.expand_dims(image, axis=-1), np.expand_dims(mask, axis=-1)

    def __getitem__(self, index):
        batch_ids = self.ids[index * self.batch_size : (index + 1) * self.batch_size]
        images, masks = zip(*(self.__load__(i) for i in batch_ids))
        return [np.zeros_like(np.array(images)), np.array(images)], np.array(masks)

    def on_epoch_end(self):
        if self.shuffle:
            random.shuffle(self.ids)
        gc.collect()

    def __len__(self):
        return int(np.ceil(len(self.ids) / self.batch_size))

# ===========================
# Loss & Metrics
# ===========================
def dice_coef(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    return 1. - dice_coef(y_true, y_pred)

def iou_score(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    union = tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

def total_loss_fn(shape_prior, alpha=0.1):
    prior_tensor = tf.convert_to_tensor(shape_prior, dtype=tf.float32)
    def loss(y_true, y_pred):
        d_loss = dice_loss(y_true, y_pred)
        s_loss = tf.reduce_mean(tf.square(y_pred - prior_tensor))
        return d_loss + alpha * s_loss
    return loss

def compute_shape_prior_from_generator(generator):
    all_masks = []
    for i in range(len(generator)):
        _, masks = generator[i]
        all_masks.append(masks)
    all_masks = np.concatenate(all_masks, axis=0)
    return np.mean(all_masks, axis=0)

# ===========================
# Paths
# ===========================
train_img_path = "/workspace/final_split_balanced/final_split_balanced/augmentedtrain/images"
train_mask_path = "/workspace/final_split_balanced/final_split_balanced/augmentedtrain/masks"
valid_img_path = "/workspace/final_split_balanced/final_split_balanced/valid/images"
valid_mask_path = "/workspace/final_split_balanced/final_split_balanced/valid/masks"

train_image_ids = sorted(os.listdir(train_img_path))
valid_image_ids = sorted(os.listdir(valid_img_path))

train_gen = DICOMDataGen(train_image_ids, train_img_path, train_mask_path, shuffle=True)
valid_gen = DICOMDataGen(valid_image_ids, valid_img_path, valid_mask_path, shuffle=False)

# ===========================
# 1️⃣ Build UNet++ (Brain MRI) to load pretrained weights
# ===========================
unetpp_path = "/workspace/UNetPP (1).py"
spec_unetpp = importlib.util.spec_from_file_location("UNetPP", unetpp_path)
unetpp_module = importlib.util.module_from_spec(spec_unetpp)
spec_unetpp.loader.exec_module(unetpp_module)
UNetPlusPlus = unetpp_module.UNetPlusPlus

pretrained_unetpp = UNetPlusPlus(input_shape=(img_size, img_size, 1)).BuildNetwork()
pretrained_unetpp.load_weights("/workspace/UNetpp_BrainTumorSegment.h5")

# ===========================
# 2️⃣ Build DualEncoderUNet and transfer Encoder A weights
# ===========================
model = build_dual_encoder_unet(input_shape=(img_size, img_size, 1), filters=filters)

layer_map = {
    "conv00_1": "A_conv00_conv1",
    "conv00_2": "A_conv00_conv2",
    "conv10_1": "A_conv10_conv1",
    "conv10_2": "A_conv10_conv2",
    "conv20_1": "A_conv20_conv1",
    "conv20_2": "A_conv20_conv2",
    "conv30_1": "A_conv30_conv1",
    "conv30_2": "A_conv30_conv2",
}
for old_name, new_name in layer_map.items():
    pretrained_w = pretrained_unetpp.get_layer(old_name).get_weights()
    model.get_layer(new_name).set_weights(pretrained_w)
    model.get_layer(new_name).trainable = False

# ===========================
# Compile
# ===========================
shape_prior = compute_shape_prior_from_generator(train_gen)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
    loss=total_loss_fn(shape_prior, alpha=0.1),
    metrics=[dice_coef, iou_score]
)

# ===========================
# Train
# ===========================
callbacks = [
    EarlyStopping(monitor='val_dice_coef', mode='max', patience=20, restore_best_weights=True, verbose=1)
]
history = model.fit(
    train_gen,
    validation_data=valid_gen,
    epochs=epochs,
    callbacks=callbacks,
    verbose=1
)

# ===========================
# Plot curves
# ===========================
history_dict = history.history
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.plot(history_dict['loss'], label='Train Loss')
plt.plot(history_dict['val_loss'], label='Val Loss')
plt.title('Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history_dict['dice_coef'], label='Train Dice')
plt.plot(history_dict['val_dice_coef'], label='Val Dice')
plt.title('Dice Coefficient Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Dice')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(history_dict['iou_score'], label='Train IoU')
plt.plot(history_dict['val_iou_score'], label='Val IoU')
plt.title('IoU Score Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('IoU')
plt.legend()

plt.tight_layout()
plt.show()

# ===========================
# Save model
# ===========================
model.save("/workspace/dual_encoder_shapeprior_trained.h5")
print("✅ Model saved to /workspace/dual_encoder_shapeprior_trained.h5")

tf.keras.backend.clear_session()
gc.collect()
```

```python
# ===========================
# Hyperparameters
# ===========================
img_size = 512
batch_size = 4
warmup_epochs = 5
epochs = 50
lr_frozen = 1e-4
lr_unfrozen = 1e-5
alpha_shape_prior = 0.05
filters = [32, 64, 128, 256, 512]

import os
import numpy as np
import tensorflow as tf
import importlib.util
import matplotlib.pyplot as plt
import cv2
import random
import gc
from tensorflow.keras.callbacks import EarlyStopping

# ===========================
# Load Adapted DualEncoder
# ===========================
dual_encoder_path = "/workspace/dual_encoder_with_shared_tail_adapted.py"
spec = importlib.util.spec_from_file_location("DualEncoderSharedTail_Adapted", dual_encoder_path)
dual_module = importlib.util.module_from_spec(spec)
spec.loader.exec_module(dual_module)
build_dual_encoder_adapted = dual_module.build_dual_encoder_with_shared_tail_adapted

# ===========================
# Data Generator
# ===========================
class DICOMDataGen(tf.keras.utils.Sequence):
    def __init__(self, image_ids, img_path, mask_path, batch_size=batch_size, image_size=img_size, shuffle=True):
        self.ids = image_ids
        self.img_path = img_path
        self.mask_path = mask_path
        self.batch_size = batch_size
        self.image_size = image_size
        self.shuffle = shuffle
        self.on_epoch_end()

    def __load__(self, item):
        image = cv2.imread(os.path.join(self.img_path, item), cv2.IMREAD_GRAYSCALE).astype(np.float32) / 255.0
        mask = cv2.imread(os.path.join(self.mask_path, item), cv2.IMREAD_GRAYSCALE).astype(np.float32)
        mask = (mask >= 255).astype(np.float32)
        image = cv2.resize(image, (self.image_size, self.image_size), interpolation=cv2.INTER_LINEAR)
        mask = cv2.resize(mask, (self.image_size, self.image_size), interpolation=cv2.INTER_NEAREST)
        return np.expand_dims(image, axis=-1), np.expand_dims(mask, axis=-1)

    def __getitem__(self, index):
        batch_ids = self.ids[index * self.batch_size : (index + 1) * self.batch_size]
        images, masks = zip(*(self.__load__(i) for i in batch_ids))
        return np.array(images), np.array(masks)

    def on_epoch_end(self):
        if self.shuffle:
            random.shuffle(self.ids)
        gc.collect()

    def __len__(self):
        return int(np.ceil(len(self.ids) / self.batch_size))

# ===========================
# Loss & Metrics
# ===========================
def dice_coef(y_true, y_pred, smooth=1e-6):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)
    y_true = tf.image.resize(y_true, tf.shape(y_pred)[1:3], method='nearest')
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

def iou_score(y_true, y_pred, smooth=1e-6):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)
    y_true = tf.image.resize(y_true, tf.shape(y_pred)[1:3], method='nearest')
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    union = tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

def dice_loss(y_true, y_pred):
    return 1. - dice_coef(y_true, y_pred)

def total_loss_fn(shape_prior, alpha=0.1):
    prior_tensor = tf.convert_to_tensor(shape_prior, dtype=tf.float32)
    def loss(y_true, y_pred):
        y_true = tf.cast(y_true, tf.float32)
        y_pred = tf.cast(y_pred, tf.float32)
        d_loss = dice_loss(y_true, y_pred)
        prior_resized = tf.image.resize(prior_tensor, tf.shape(y_pred)[1:3], method='nearest')
        prior_resized = tf.expand_dims(prior_resized, axis=0)
        prior_resized = tf.repeat(prior_resized, tf.shape(y_pred)[0], axis=0)
        s_loss = tf.reduce_mean(tf.square(y_pred - prior_resized))
        return d_loss + alpha * s_loss
    return loss

def compute_shape_prior_from_generator(generator, cache_path="shape_prior.npy"):
    if os.path.exists(cache_path):
        print(f"Loading cached shape prior from {cache_path}")
        return np.load(cache_path)
    all_masks = []
    for i in range(len(generator)):
        _, masks = generator[i]
        all_masks.append(masks)
    all_masks = np.concatenate(all_masks, axis=0)
    prior = np.mean(all_masks, axis=0)
    np.save(cache_path, prior)
    print(f"Saved shape prior to {cache_path}")
    return prior

# ===========================
# Paths
# ===========================
train_img_path = "/workspace/final_split_balanced/final_split_balanced/augmentedtrain/images"
train_mask_path = "/workspace/final_split_balanced/final_split_balanced/augmentedtrain/masks"
valid_img_path = "/workspace/final_split_balanced/final_split_balanced/valid/images"
valid_mask_path = "/workspace/final_split_balanced/final_split_balanced/valid/masks"

train_image_ids = sorted(os.listdir(train_img_path))
valid_image_ids = sorted(os.listdir(valid_img_path))

train_gen = DICOMDataGen(train_image_ids, train_img_path, train_mask_path, shuffle=True)
valid_gen = DICOMDataGen(valid_image_ids, valid_img_path, valid_mask_path, shuffle=False)

# ===========================
# Load UNet++ (MRI weights for Encoder A)
# ===========================
unetpp_path = "/workspace/UNetPP (1).py"
spec_unetpp = importlib.util.spec_from_file_location("UNetPP", unetpp_path)
unetpp_module = importlib.util.module_from_spec(spec_unetpp)
spec_unetpp.loader.exec_module(unetpp_module)
UNetPlusPlus = unetpp_module.UNetPlusPlus

pretrained_unetpp = UNetPlusPlus(input_shape=(img_size, img_size, 1)).BuildNetwork()
pretrained_unetpp.load_weights("/workspace/UNetpp_BrainTumorSegment.h5")

# ===========================
# Build new adapted model
# ===========================
model = build_dual_encoder_adapted(input_shape=(img_size, img_size, 1), filters=filters)

# Weight mapping for Encoder A
layer_map = {
    "conv00_1": "A_conv00_conv1",
    "conv00_2": "A_conv00_conv2",
    "conv10_1": "A_conv10_conv1",
    "conv10_2": "A_conv10_conv2",
    "conv20_1": "A_conv20_conv1",
    "conv20_2": "A_conv20_conv2",
    "conv30_1": "A_conv30_conv1",
    "conv30_2": "A_conv30_conv2",
    "conv40_1": "A_conv40_conv1",
    "conv40_2": "A_conv40_conv2",
}

# Load MRI weights into Encoder A
print("Loading MRI weights into Encoder A:")
for old_name, new_name in layer_map.items():
    pretrained_w = pretrained_unetpp.get_layer(old_name).get_weights()
    model.get_layer(new_name).set_weights(pretrained_w)
    model.get_layer(new_name).trainable = False
    print(f"  → Loaded & froze {new_name}")

# ===========================
# Check input to Encoder A
# ===========================
sample_x, _ = train_gen[0]
print(f"Sample CT batch shape: {sample_x.shape}")
print(f"Adapted input to Encoder A stats — min: {sample_x.min()}, max: {sample_x.max()}, mean: {sample_x.mean()}")

# ===========================
# Shape Prior
# ===========================
shape_prior = compute_shape_prior_from_generator(train_gen)

# ===========================
# Warmup Phase
# ===========================
print("🔹 Warmup: Encoder A frozen, training Encoder B + Decoder")
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=lr_frozen),
    loss=dice_loss,
    metrics=[dice_coef, iou_score]
)
model.fit(
    train_gen,
    validation_data=valid_gen,
    epochs=warmup_epochs,
    verbose=1
)

# ===========================
# Fine-tuning Phase
# ===========================
print("🔹 Fine-tuning: Unfreezing Encoder A")
for old_name, new_name in layer_map.items():
    model.get_layer(new_name).trainable = True
    print(f"  → Unfroze {new_name}")

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=lr_unfrozen),
    loss=total_loss_fn(shape_prior, alpha=alpha_shape_prior),
    metrics=[dice_coef, iou_score]
)

callbacks = [
    EarlyStopping(monitor='val_dice_coef', mode='max', patience=15, restore_best_weights=True, verbose=1)
]
history = model.fit(
    train_gen,
    validation_data=valid_gen,
    epochs=epochs,
    callbacks=callbacks,
    verbose=1
)

# ===========================
# Plot curves
# ===========================
history_dict = history.history
plt.figure(figsize=(18, 5))
plt.subplot(1, 3, 1)
plt.plot(history_dict['loss'], label='Train Loss')
plt.plot(history_dict['val_loss'], label='Val Loss')
plt.title('Loss')
plt.legend()
plt.subplot(1, 3, 2)
plt.plot(history_dict['dice_coef'], label='Train Dice')
plt.plot(history_dict['val_dice_coef'], label='Val Dice')
plt.title('Dice')
plt.legend()
plt.subplot(1, 3, 3)
plt.plot(history_dict['iou_score'], label='Train IoU')
plt.plot(history_dict['val_iou_score'], label='Val IoU')
plt.title('IoU')
plt.legend()
plt.tight_layout()
plt.show()

# ===========================
# Save
# ===========================
model.save("/workspace/dual_encoder_sharedtail_adapted_trained_adpted_adapted.h5")
print("✅ Saved /workspace/dual_encoder_sharedtail_adapted_trained.h5")
```
